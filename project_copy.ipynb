```python
X = np.arange(pi/64, 2*pi + pi/64, pi/64).reshape (1,128)
Y = square(sin(X))
plt.plot(X.T, Y.T)
plt.show()
```

```python
layers_dims = [X.shape[0], 32, 32, 32, 32, 32, Y.shape[0]]
```

```python
parameters_guess = model_optimize(X, Y, layers_dims, "entropy" ,"adam", learning_rate = 0.001, beta = 0.9,
          beta1 = 0.9, beta2 = 0.999,  epsilon = 1e-8, num_epochs = 2000, print_cost = True)
```

```python
y_predict, caches = L_model_forward(X, parameters_guess)
plt.plot(X.T, Y.T, ".", label= "expected output")
plt.plot(X.T, y_predict.T, label= "output")
plt.xlabel("Input (x)")
plt.ylabel("Output (f(x))")
plt.title("Output of NN")
plt.legend ()
plt.show()
```

```python
rho_list = [3.5, 4, 4.5, 5]
output = {}
i = 0
for rho in rho_list:
    parameters = model_optimize_free_e (X, rho_list[i], matrix, parameters_guess, layers_dims, "adam", learning_rate = 0.001, beta = 0.9, beta1 = 0.9, beta2 = 0.999, epsilon = 1e-8, num_epochs = 20000, print_cost = True)
    output['rho'+ str(i)], caches = L_model_forward(X, parameters)
    normal_factor = 1/rho
    output['rho'+ str(i)] = output['rho'+ str(i)] * (normal_factor/integrate.trapz(output['rho'+ str(i)],X))
    i+=1
```
